```python
import pandas as pd
path="D:\\Avi\\College\\INTERNSHIP\\2023_CSE_Project\\Logistic\\Titanic survived or not\\train.csv"
data=pd.read_csv(path)
print(data)
print(data.info())
print(data.isnull().sum())

import sklearn
from sklearn.preprocessing import LabelEncoder
label_Gender=LabelEncoder()
data['Sex']=label_Gender.fit_transform(data['Sex'])
data['Embarked']=label_Gender.fit_transform(data['Embarked'])
data['Ticket']=label_Gender.fit_transform(data['Ticket'])
data['Cabin']=label_Gender.fit_transform(data['Cabin'])
data['Name']=label_Gender.fit_transform(data['Name'])
print(data)

median=data.Age.median()
data.Age=data.Age.fillna(median)

median=data.Cabin.median()
data.Cabin=data.Cabin.fillna(median)

median=data.Embarked.median()
data.Embarked=data.Embarked.fillna(median)

print(data.isnull().sum())


inputs=data.drop(['PassengerId','Survived'],axis=1)
outputs=data.drop(['PassengerId','Pclass','Name','Sex','Age','SibSp','Parch','Ticket','Fare','Cabin','Embarked'],axis=1)
print(inputs)
print(outputs)
```

```python
import sklearn
import pandas as pd
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
inputs=sc.fit_transform(inputs)
```

```python
import sklearn
import sklearn
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.fit(inputs,outputs)
res=model.predict([[1 ,  190  ,  0 , 38.0     , 1    , 0   ,  596  ,71.2833  ,   81 ,        0]])
print(res)
if res==1:
    print("SURVIVED")
else:
    print("DID NOT SURVIVE")
acc=model.score(inputs,outputs)
print(acc*100)
```
